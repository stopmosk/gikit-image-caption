```python
# FASTTEXT_WIKI_URL = (
#     "https://dl.fbaipublicfiles.com/pythia/pretrained_models/fasttext/wiki.en.bin"
# )

import requests
from tqdm import tqdm
from fasttext import load_model
import numpy as np
```

```python
class WordToVectorDict:
    def __init__(self, model):
        self.model = model

    def __getitem__(self, word):
        # Check if mean for word split needs to be done here
        return np.mean([self.model.get_word_vector(w) for w in word.split(" ")], axis=0)
```

```python
model = load_model('../../../../pretrained_models/wiki.en.bin')
# String to Vector
stov = WordToVectorDict(model)
```
